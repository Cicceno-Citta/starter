## 1. 初めてのプログラム

コードブロック左側の「再生ボタン」をクリックすると、コードが実行できます。（初回は少し時間がかかります）

`print("文字列")`を使うと「文字列」を出力できます。

```python
print("Hello, World!")
```

好きな文字列を出力するように編集してみよう。

## 2. 計算をしてみよう

- **変数**
    - データを格納するための箱
- **代入**
    - 変数に値を入れること
    - Pythonでは`(変数) = (値)`で表す




```python
x = 100
y = 200

print(x+y)
```

新しい変数`z`を作り、`x+y+z`を計算してみよう。

## 3. プログラム実行時に値を入力する

②だとプログラム内で指定した値しか計算できない。

`(変数) = int(input())`で変数に数値を入力することができる。

```python
x = int(input())

print(x)
```

## 4. 実行時に入力された値を計算する

②, ③を組み合わせると実行時に入力された値を計算できる。

2個の整数`x, y`を入力して、その和を出力してみよう。
